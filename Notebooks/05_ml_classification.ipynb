# 05 · ML Classification

Train and compare classifiers to discriminate between the three disease states.

Steps:
1. **RFECV feature selection** — recursive feature elimination with cross-validation
2. **Model training** — Logistic Regression, Random Forest, SVM, XGBoost, LightGBM
3. **Evaluation** — stratified 5-fold cross-validation, accuracy and macro AUC (target > 0.85)
4. **Hyperparameter tuning** — grid search on best-performing model
5. **SHAP interpretability** — feature importance and individual predictions
6. **Learning curves** — assess overfitting / data sufficiency

**Input:** `files/data/processed/normalized_matrix.pkl`  
**Output:** `results/model_results.pkl`, figures in `figures/`
