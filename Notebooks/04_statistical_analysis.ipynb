# 04 · Statistical Analysis

Explore group structure and identify statistically significant features.

Unsupervised:
- PCA (with scree plot)
- UMAP
- Hierarchical clustering / dendrogram

Supervised / univariate:
- Kruskal-Wallis test across all three groups + Benjamini-Hochberg FDR correction
- Pairwise volcano plots (cancer vs. healthy, cancer vs. benign)
- PLS-DA
- VIP scores

**Input:** `files/data/processed/normalized_matrix.pkl`  
**Output:** `results/statistical_results.pkl`, figures in `figures/`
